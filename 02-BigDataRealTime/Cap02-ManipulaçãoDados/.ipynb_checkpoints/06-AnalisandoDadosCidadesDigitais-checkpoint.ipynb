# Group By e Reshape

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html

```python
import numpy as np
import pandas as pd
```

```python
cidadesDigitais = pd.read_csv('cidades_digitais.csv', sep = ';', encoding = 'latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.count()
```

```python
del cidadesDigitais['IBGE']
```

```python
# Aplicando Group By
cidadesUF_1 = cidadesDigitais.groupby('UF')
```

```python
type(cidadesUF_1)
```

```python
cidadesUF_1.groups
```

```python
# Número de grupos
len(cidadesUF_1.groups)
```

```python
ordenaUF = cidadesUF_1.size()
```

```python
# Ordena os grupos por ordem decrescente
# Minha Resposta = DSA
ordenaUF.sort_values(ascending=False)
```

```python
cidadesUF_2 = cidadesDigitais.groupby(['UF','STATUS'])
```

```python
help(cidadesUF_2.count)
```

```python
# Solução sem GroupBy
pd.pivot_table(cidadesDigitais, values='CIDADE', index=['UF', 'STATUS'], aggfunc=np.count_nonzero)
```

```python
# Minha Resposta
print(cidadesUF_2.count()[['CIDADE']])
```

```python
# Resppsta II
ordenaUF = cidadesUF_2.count()
print(cidadesUF_2.groups)
```

```python
ordenaUF.sort_index(ascending=True)
```

```python
# Total agrupado
# DSA
total = cidadesUF_2.size()
total.sort_values(ascending=False)
total
```

```python
# Sem alteração no original, além da remoção da coluna IBGE
cidadesDigitais.head()
```

```python
cidadesUF_3 = cidadesDigitais.set_index('STATUS')
```

```python
# Aplicando Group By
# Minha Resposta
cidadesUF_3 = cidadesUF_3.groupby('STATUS')
```

```python
cidadesUF_3.head(5)
```

```python
# DSA
cidadesUF_3 = cidadesDigitais.set_index('STATUS')
cidadesUF_3 = cidadesUF_3.groupby(level = 'STATUS')
```

```python
cidadesUF_3.head()
```

```python
# Solução para cidadeUF_4
cidadesUF_4 = cidadesDigitais.set_index(['STATUS', 'UF']).groupby(['UF', 'CIDADE'])
```

```python
cidadesUF_4.head()
```

```python
# DSA
cidadeUF_4 = cidadesDigitais.set_index(['STATUS', 'UF'])
```

```python
cidadesUF_4
```

```python
# 
cidadesUF_2['POPULAÇÃO', 'PONTOS_ATENDIDOS'].sum()
    
```

```python
# Alternativa ao warining
cidadesUF_2[['POPULAÇÃO', 'PONTOS_ATENDIDOS']].sum() # -- apply(sum)
```

```python
# DSA
cidadesUF_4 = cidadesDigitais.set_index(['UF', 'STATUS'])
cidadesUF_4 = cidadesUF_4.groupby(level = ['UF', 'STATUS'])
cidadesUF_4.sum()
```

```python
# Aparentemente, o mesmo que o anterior
cidadesUF_4.aggregate(np.sum)
```

```python
# Solução
pd.pivot_table(cidadesDigitais, 
               index=['UF', 'STATUS'], 
               aggfunc=[np.sum, np.mean, np.size],
               values=['POPULAÇÃO', 'PONTOS_ATENDIDOS'],                                
              )
```

```python
# DSA
cidadesUF_4.aggregate([np.sum, np.mean, np.size])
```

```python
# Solução
cidadesUF_5 = cidadesDigitais.set_index(['STATUS', 'CIDADE', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO', 'POPULAÇÃO', 'PONTOS_ATENDIDOS'])
cidadesUF_5.head()
```

```python
cidadesUF_5 = cidadesUF_5.groupby(['STATUS', 'VALOR_TOTAL_PREVISTO', 'VALOR_INVESTIDO'])
```

```python
cidadesUF_5.sum()
```
