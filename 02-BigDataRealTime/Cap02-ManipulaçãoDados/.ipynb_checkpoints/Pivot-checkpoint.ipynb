# Reshaping and Pivot Tables

```python
import pandas as pd
import numpy as np
import pandas._testing as tm

# Criação da tabela a ser utilizada no exemplo
def unpivot(frame):
    N, K = frame.shape
    data = {
        "value": frame.to_numpy().ravel("F"),
        "variable": np.asarray(frame.columns).repeat(N),
        "date": np.tile(np.asarray(frame.index), K),
    }
    return pd.DataFrame(data, columns=["date", "variable", "value"])


df = unpivot(tm.makeTimeDataFrame(3))
```

```python
df
```

## Selecionar apenas os registros de variável A

```python
df[df["variable"] == "A"]
```

## Tranformar alguns registros da tabela em colunas

```python
df.pivot(index="date", columns="variable", values="value")
```

## Adicionar a coluna 

```python
df['value2'] = df['value'] * 2
```

```python
df
```

```python
pivoted = df.pivot(index = 'date', columns='variable')
```

```python
pivoted
```

## subset

```python
pivoted['value2']
```

# Reshaping por stacking 2 unstacking

```python
tuples = list(
    zip(
        *[
            ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
            ["one", "two", "one", "two", "one", "two", "one", "two"],
        ]
    )
)
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=["first", "second"])
```

```python
df = pd.DataFrame(np.random.randn(8,2),
                 index = index,
                 columns=["A", "B"])
```

```python
df
```

```python
df2 = df[:4]
```

```python
df2
```

```python
stacked = df2.stack()
```

```python
display(stacked)
```

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

```python
stacked.unstack("second")
```

# Multi-Index

```python
index = pd.MultiIndex.from_product([[2, 1], ['a', 'b']])
```

```python
df = pd.DataFrame(np.random.rand(4),
                 index=index,
                 columns=["A"])
```

```python
df
```

```python
all(df.unstack().stack() == df.sort_index())
```

## Multiple levels

```python
columns = pd.MultiIndex.from_tuples(
    [
        ("A", "cat", "long"),
        ("B", "cat", "long"),
        ("A", "dog", "short"),
        ("B", "dog", "short"),        
    ],
    names = ["exp", "animal", "hair_length"],
)
```

```python
df = pd.DataFrame(np.random.randn(4,4), columns=columns)
```

```python
df
```

```python
df.stack(level=['animal', 'hair_length'])
```

```python
df.stack(level=[1, 2])
```

## Missing data

```python
columns = pd.MultiIndex.from_tuples(
    [
        ("A", "cat"),
        ("B", "dog"),
        ("B", "cat"),
        ("A", "dog"),
    ],
    names = ["exp", "animal"],
)
```

```python
index = pd.MultiIndex.from_product(
    [
        ("bar", "baz", "foo", "qux"),
        ("one", "two")
    ],
    names = ["first", "second"]
)
```

```python
df = pd.DataFrame(np.random.randn(8, 4),
                  index = index,
                  columns = columns
                 )
```

```python
df2 = df.iloc[[0, 1, 2, 4, 5, 7]]
```

```python
df2
```

```python
df2.stack("exp")
```

```python
df2.stack("animal")
```

```python
df
```

```python
df3 = df.iloc[[0, 1, 4, 7], [1, 2]]
```

```python
df3
```

```python
df3.unstack()
```

```python
df3.unstack(fill_value=-1e9)
```

# Reshaping by Melt

```python
cheese = pd.DataFrame(
    {
        "first": ["John", "Mary"],
        "last": ["Doe", "Bo"],
        "height": [5.5, 6.0],
        "weight": [130, 150],
    }
)
```

```python
cheese
```

```python
cheese.melt(id_vars=["first", "last"])
```

```python
cheese.melt(id_vars=["first", "last"], var_name="quantity")
```

```python
index = pd.MultiIndex.from_tuples(
    [
        ("person", "A"),
        ("person", "B")
    ]
)
```

```python
cheese = pd.DataFrame(
    {
        "first": ["John", "Mary"],
        "last": ["Doe", "Bo"],
        "height": [5.5, 6.0],
        "weight": [130, 150],
    },
    index = index,
)
```

```python
cheese
```

```python
cheese.melt(id_vars=['first', 'last'])
```

```python
cheese.melt(id_vars=['first', 'last'], ignore_index = False)
```
